```python
import json
import matplotlib.pyplot as plt

%matplotlib inline
```

## AutoEncoder training process

```python
with open('./ckpt/SGD_0.1_MSE_batch100_100epoch_div8_unbalanced/loss.json') as f:
  losses = json.loads(f.read())

plt.xlabel('Epochs', fontsize=10)
plt.ylabel('Loss', fontsize=10)

# plt.xticks(range(0, 500000, 50000),fontsize=8)
plt.yticks(fontsize=8)

plt.plot(losses['loss'])
```

## Classifier training process

```python
with open('./ckpt/test_cls_SGD_0.1_MSE_batch100_100epoch_div8_unbalanced/loss.json') as f:
  losses = json.loads(f.read())

plt.xlabel('Epochs', fontsize=10)
plt.ylabel('Loss', fontsize=10)

# plt.xticks(range(0, 500000, 50000),fontsize=8)
plt.yticks(fontsize=8)

plt.plot(losses['train_loss'])
plt.plot(losses['test_loss'])
```

```python
plt.xlabel('Epochs')
plt.ylabel('Acc')

plt.plot(losses['train_acc'])
plt.plot(losses['test_acc'])
```
